## Lecture des donn√©es V√©lib

```python
# chargement des donn√©es de status et d'information depuis l'API V√©lib' M√©tropole (fonctions supports)
#
Status = tools.loadVelibStatus()
Information = tools.loadVelibInformation()

Status = tools.getVelibStations(Status)
Information = tools.getVelibStations(Information)
```

## Pr√©paration des donn√©es

```python
# conversion en Data Frames Pandas des donn√©es de status et d'information (pd.DataFrame)
Status_df = pd.DataFrame(Status)
Information_df = pd.DataFrame(Information)

# fusion des deux Data Frames (fonction merge)
Velib_df = pd.merge(Status_df, Information_df)

# affichage du r√©sultat
display(Velib_df.head())
```

## Analyses statistiques

```python
# mesures statistiques assembl√©es (fonctions describe et join)
Statistics1 = Status_df.describe()
Statistics2 = Information_df.describe()

# Perform the join
Statistics = Statistics1.join(Statistics2,rsuffix='_status',lsuffix='_info')
# Statistics3 = Statistics2.join(Statistics1,rsuffix='_info',lsuffix='_status')

#
# affichage du r√©sultat
display(Statistics)

# display(Statistics3)

# Mesures statistiques sur les donn√©es V√©lib avec somme
# Statistics = Velib_df.describe(include='all').append(Velib_df.sum().rename('sum'))
columns_to_sum = Velib_df.columns[Velib_df.dtypes != 'object']  # Exclude non-numeric columns
Statistics = Velib_df[columns_to_sum].describe(include='all')
Statistics.loc['sum'] = Velib_df[columns_to_sum].sum()

# Affichage du r√©sultat
display(Statistics)

```

## Exportation pour le Web

```python
# renommage des colonnes avant l'exportation en HTML (fonction rename)
Statistics = Statistics.rename(columns={'stationCode': 'stationCode', 'numBikesAvailable': 'numBikesAvailable', 'numBikesAvailable': 'numBikesAvailable', 'numBikesAvailable': 'numBikesAvailable', 'numBikesAvailable': 'numBikesAvailable', 'numDocksAvailable': 'numDocksAvailable', 'numDocksAvailable': 'numDocksAvailable', 'numDocksAvailable': 'numDocksAvailable', 'numDocksAvailable': 'numDocksAvailable', 'isInstalled': 'isInstalled', 'isReturning': 'isReturning', 'isRenting': 'isRenting', 'lastReported': 'lastReported', 'latitude': 'latitude', 'longitude': 'longitude', 'stationCapacity': 'stationCapacity'})


# exportation des mesures au format HTML (fonctions open et to_html)
f = open('/content/drive/MyDrive/Colab Notebooks/SAE15/tests/Statistics.html', 'w')
f.write(Statistics.to_html())
f.close()

```
